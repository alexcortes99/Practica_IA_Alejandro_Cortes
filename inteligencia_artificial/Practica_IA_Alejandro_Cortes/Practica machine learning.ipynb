## Normalización y ajustes de las unidades

### foot_float
El método **foot_float** convierte el valor de la altura en un valor float. Es decir, si tenemos un valor de altura igual a 5'2 pies, la función lo convierte en 5.2 y por lo tanto procesable para la regresión.

```python
def foot_float(x):
    if "'" in x:
        ret_val = float(x.replace("'", "."))
    return ret_val
```

```python
df["Height"] = df["Height"].apply(foot_float)
df.head()
```

### remove_lbs

Esta función quita las letras `lbs` de las unidades de la columna `Weight`.

```python
def remove_lbs(x):
    ret_val=0.0
    if 'lbs' in x:
        if len(x) > 1:
            ret_val = float(x.replace('lbs', ''))
    return ret_val
```

```python
df["Weight"] = df["Weight"].apply(remove_lbs)
df.head()
```

### Quitar los NaNs de los porteros

Para quitar los valores `NaNs` de los porteros se ha optado por sustituir (imputar) los NaNs por una métrica de agregación del resto de jugadores
Para ello se ha creado una función que detecte si el valor es `NaN`. Luego se ha calculado la media para cada columna en la que tienen `NaN` los porteros (los modificadores de posición).

```python
def isNaN(num):
    return num != num
def parse_stats(x):
    if isNaN(x):
        return x
    if "+" in x:
        ret_val_aux=x.split("+")
        ret_val= int(ret_val_aux[0]) + int(ret_val_aux[1])
    return ret_val
```

### Value_to_float

Para los valores monetarios se ha utilizado la función `Value_to_float` que pasa las unidades a euros multiplicando las `K` por mil y las `M` por un millón. Las columnas afectadas por esta función son `Value`, `Wage`, `Release Clause`.

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
df.head()
```

## Entrenamiento de los datos y predicción

Para entrenar los datos escogemos primero la variable a estudiar `Value`. Luego dividimos los datos con la función **train_test_split**. En este caso se considerará un 33% de datos de testeo y un 66% de entrenamiento. 

```python
val = df.pop("Value")
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=42)
```

```python
len(X_train)
```

Utilizamos los datos para realizar un modelo de regresión lineal y lanzamos una predicción. Tanto la función **LinearRegression** como la función **predict** son de la libreria `scikit-learn` 

```python
reg = linear_model.LinearRegression().fit(X_train, y_train) 
```

```python
preds = reg.predict(X_test)
```

Si nos fijamos en el primer valor real, se trata de 5 millones mientras que el primer valor predecido es de 4,7 millones, un valor relativamente parecido si tenemos en cuenta la magnitud de valores que estamos tratando.

```python
r2_score(preds, y_test)
```

Como se puede apreciar, el r2 nos ha salido de 0,993. Aunque sea un valor bastante alto no significa que sea óptimo. Esto se trata de **overfitting**, es decir, se ha entrenado demasiado los datos de tal manera que tendremos un porcentaje de acierto muy alto y si se nos presentara un dataframe totalmente distinto a lo entrenado es bastante probable que no haga las predicciones adecuadas.
